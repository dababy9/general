```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# read main csv
df = pd.read_csv('datasets/arcos-sc.csv', low_memory=False)

# The following was written with help from ChatGPT

# convert column containing dates to actual datetime objects
df["TRANSACTION_DATE"] = pd.to_datetime(df["TRANSACTION_DATE"])

# now set the column to just contain year of transaction
df["TRANSACTION_DATE"] = df["TRANSACTION_DATE"].dt.year

# function to format matplotlib graphs without exponential numbers
def format(value, tick_number):
    return f'{value:.0f}'
```

#### 1. How many pills were sold in this state in the whole dataset?

```python
# sum up Series for 'DOSAGE_UNIT' column (which is # of pills for each order)
print(df["DOSAGE_UNIT"].sum())
```

#### 2. Using matplotlib, make a line graph, and then a bar graph showing the number of pills bought in each year of 2006-2014 (same data in two different forms).

```python
# group DataFrame by year and sum pill counts across each year
ppy = df.groupby('TRANSACTION_DATE')["DOSAGE_UNIT"].sum().reset_index()

# only retain years 2006-2014
ppy = ppy.query('TRANSACTION_DATE >= 2006 and TRANSACTION_DATE <= 2014')

# bar graph of pills per year
plt.bar(ppy["TRANSACTION_DATE"], ppy["DOSAGE_UNIT"])

# format graph
plt.gca().yaxis.set_major_formatter(FuncFormatter(format))
```

```python
# line graph of pills per year
plt.plot(ppy["TRANSACTION_DATE"], ppy["DOSAGE_UNIT"])

# format graph
plt.gca().yaxis.set_major_formatter(FuncFormatter(format))
```

#### 3. Which five counties bought the most pills?

```python
# I enjoy making one liners...
# groupby() returns a iterable object of dataframes, grouped by having the same 'BUYER_COUNTY' value
# ["DOSAGE_UNIT"] specifies only the pill count to be looked at
# sum() just sums up the pill counts for each order, for each county, and returns a Series with the keys as counties and the values as total pill counts
# sort_values(ascending=False) sorts the series by total pill counts (ascending to have the largest pill counts at the top)
# head(5) grabs the top 5 rows of the Series
# index.values just grabs the keys of the top 5 rows and then returns the array of keys
print(df.groupby('BUYER_COUNTY')["DOSAGE_UNIT"].sum().sort_values(ascending=False).head(5).index.values)
```

#### 4. Which five counties bought the most pills per capita?

```python
# read county populations into DataFrame
pop = pd.read_csv('datasets/populations.csv')

# extract DataFrame with counties and total pill counts
pills = df.groupby('BUYER_COUNTY')["DOSAGE_UNIT"].sum().to_frame().reset_index()

# merge new population Series with grouped DataFrame
merged = pd.merge(pop, pills, on='BUYER_COUNTY')

# create new column by dividing pill count by population (pills per capita)
merged['PILLS_PER_CAPITA'] = merged['DOSAGE_UNIT'] / merged['population']

# return top 5 counties sorted by the new column
print(merged.sort_values(by='PILLS_PER_CAPITA', ascending=False)['BUYER_COUNTY'].head(5).values)
```

Ask a question along these lines (do more religious communities use fewer pain pills?), and partially answer it by plotting the variable of your choice against per-capita pain pill usage.
```python
# read unemployment data for each county from 2012-2022
unpt = pd.read_csv('datasets/unemployment.csv')

# create new column with average unemployment rate for all 10 years
unpt["AVG_UNPT"] = unpt.loc[:, '2012':'2022'].sum(axis=1) / 10

# merge datasets
udf = pd.merge(unpt, merged, on='BUYER_COUNTY')

# scatter plot of average unemployment rate vs pills per capita
plt.scatter(udf["AVG_UNPT"], udf["PILLS_PER_CAPITA"]);
```
